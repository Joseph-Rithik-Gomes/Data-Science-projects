#Splitting Data

```python
year=pd.to_datetime(raw_df.Date).dt.year
train_df=raw_df[year<2015]
val_df=raw_df[year==2015]
test_df=raw_df[year>2015]
```

```python
print("train_df shape: ",train_df.shape)
print("val_df shape: ",val_df.shape)
print("test_df shape: ",test_df.shape)
```

#Recognize Input and Target Columns,and numeric and categorical columns

```python
input_cols=list(train_df.columns)[1:-1]
target_cols='RainTomorrow'
```

```python
train_inputs=train_df[input_cols]
train_target=train_df[target_cols]

val_inputs=val_df[input_cols]
val_target=val_df[target_cols]

test_inputs=test_df[input_cols]
test_target=test_df[target_cols]
```

```python
numeric_cols=list(train_inputs.select_dtypes(include=np.number).columns)
categorical_cols=list(train_inputs.select_dtypes('object').columns)
```

#Imputation

```python
from sklearn.impute import SimpleImputer
```

```python
imp=SimpleImputer(strategy='mean').fit(raw_df[numeric_cols])
```

```python
train_inputs[numeric_cols]=imp.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols]=imp.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols]=imp.transform(test_inputs[numeric_cols])
```

```python
numeric_cols
```

#Scalling Numeric Features

```python
val_inputs.describe().loc[['min','max']]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc=MinMaxScaler().fit(raw_df[numeric_cols])
```

```python
train_inputs[numeric_cols]=sc.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols]=sc.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols]=sc.transform(test_inputs[numeric_cols])
```

```python
val_inputs.describe().loc[['min','max']]
```

#Encode Categorical Data

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df=train_inputs[categorical_cols]
df.iloc[[74]]
```

```python
for i in range(len(categorical_cols)):
 train_inputs[categorical_cols[i]].fillna(train_inputs[categorical_cols[i]].mode()[0],inplace=True)
```

```python
train_inputs[categorical_cols].isna().sum()
```

```python
encoder=OneHotEncoder(sparse=False,handle_unknown='ignore').fit(raw_df[categorical_cols])
```

###Getting the names of encoded columns



```python
encoded_cols=encoder.get_feature_names_out(categorical_cols).tolist()
```

```python
train_inputs[encoded_cols]=encoder.transform(train_inputs[categorical_cols])
val_inputs[encoded_cols]=encoder.transform(val_inputs[categorical_cols])
test_inputs[encoded_cols]=encoder.transform(test_inputs[categorical_cols])
```

#Extarcting our X variables for test,val,and test

```python
x_train=train_inputs[numeric_cols+encoded_cols]
x_val=val_inputs[numeric_cols+encoded_cols]
x_test=test_inputs[numeric_cols+encoded_cols]
```

###trying different depths

```python
def try_depth(depth):
  model=DecisionTreeClassifier(max_depth=depth,random_state=42)
  model.fit(x_train,train_target)
  train_acc=1-model.score(x_train,train_target)
  val_acc=1-model.score(x_val,val_target)
  return{'Max Depth':depth, 'Training Error':np.round(train_acc*100),'Validation Error': np.round(val_acc*100)}
```

```python
errors_df=pd.DataFrame([try_depth(depth) for depth in range (1,21)])
```

```python
errors_df.sort_values('Validation Error')
```

```python
plt.figure()
plt.plot(errors_df['Max Depth'],errors_df['Training Error'],label='Training')
plt.plot(errors_df['Max Depth'],errors_df['Validation Error'],label='Validation')
plt.xlabel('Max Depth')
plt.xticks(range(0,21,2))
plt.legend()
plt.show()
```
